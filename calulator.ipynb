```python
from functions import *
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig

T = 300

def fermi(EtoEfinkT): # need to override
    return 1/(1+np.exp(EtoEfinkT))
Eg = 6
Egd2 = Eg/2
E = np.arange(-10, 10+.1, +.1) # in units of kT from Ef or eV*K with 0 being Ef
f = fermi(E)
Ev = -Egd2
Ec = Egd2
Dc = dos(si_mn, E-Ec)
Dv = dos(si_mp, Ev-E)

plt.figure()
plt.subplot(1,3,1)
plt.title("Fermi levels in kT")
plt.ylabel("E (in kT)")
plt.xlabel("probility")
plt.grid()
plt.plot(f, E, label="eletrons")
plt.plot(1-f, E, label="holes")
plt.yticks([Ev-.5, Ev, 0, Ec, Ec+.5],["Ev-KT/2","Ev", "Ef", "Ec", "Ec+KT/2"])
plt.legend()
plt.subplot(1,3,2)
plt.title("Dencity of States")
plt.plot(Dc, E, label="Dc")
plt.plot(Dv, E, label="Dv")
plt.xlabel("$N(E)$")
plt.yticks([Ev-.5, Ev, 0, Ec, Ec+.5],["Ev-KT/2","Ev", "Ef", "Ec", "Ec+KT/2"])
plt.grid()
plt.legend()
plt.subplot(1,3,3)
plt.title("Carrier Distributions")
plt.plot(f*Dc, E, label="n")
plt.plot((1-f)*Dv, E, label="p")
plt.legend()
plt.yticks([Ev-.5, Ev, 0, Ec, Ec+.5],["Ev-KT/2","Ev", "Ef", "Ec", "Ec+KT/2"])
plt.xlabel("$cm^{-3}$")
plt.grid()
plt.savefig("Carrier_Consentration_ni.png")
plt.tight_layout()

```
