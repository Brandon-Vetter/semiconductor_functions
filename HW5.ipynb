<p style='text-align: right;'> (5.5.1) </p>

$$
W_{dmax} = \sqrt{\frac{2 \epsilon_s 2 \phi_B}{qN_a}}
$$

<p style='text-align: right;'> (5.4.3) </p>

$$
V_t = V_{fb} + 2 \phi_B + \frac{\sqrt{q N_a 2 \epsilon_s 2 \phi_B}}{C_{ox}}
$$





First, lets divide eqt 5.4.3 by $\sqrt{(q Na)^2}$

$$
\frac{V_t}{q N_a} = \frac{V_{fb}}{N_a q} + \frac{2 \phi_B}{N_a q} + \sqrt{ \frac{2 \epsilon_{ox} 2 \phi_B}{q N_a}} \frac{1}{C_{ox}}
$$

Now we can sub $\sqrt{ \frac{2 \epsilon_{ox} 2 \phi_B}{q N_a}}$ for $W_{dmax}$

$$
\frac{V_t}{q N_a} = \frac{V_{fb}}{N_a q} +  \frac{2 \phi_B}{N_a q} + \frac{W_{dmax}}{C_{ox}}
$$

Now solve for $W_{dmax}$

$$
\frac{C_{ox} (V_t - V_{fb} -  2 \phi_B)}{N_a q} = W_{dmax}
$$

Also going to sub $C_{ox}$ for its equation:

$$
\frac{\epsilon_{ox} (V_t - V_{fb} -  2 \phi_B)}{N_a q T_{oxe}} = W_{dmax}
$$

Now plugging into eqt 7.3.4:

$$
l_d \propto \sqrt[3]{\frac{\epsilon_{ox} (V_t - V_{fb} -  2 \phi_B)}{N_a q}X_j}
$$
